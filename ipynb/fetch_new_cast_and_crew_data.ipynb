```python
import requests
import time
from tqdm import tqdm  # Import the tqdm library

API_URL_BEFORE = 'https://api.themoviedb.org/3/movie/' 
API_URL_AFTER = '/credits?api_key=e9e9d8da18ae29fc430845952232787c&'
requests_per_second = 50
connections_per_ip = 20
dfs = []
# Calculate the delay between requests to respect the rate limit
delay_between_requests = 1 / requests_per_second

# Use tqdm to create a progress bar for the loop
for movie_id in tqdm(movies_id, desc="Processing Movies", unit="movie"):
    # Introduce a delay between requests
    time.sleep(delay_between_requests)
        
    response = requests.get(API_URL_BEFORE + str(movie_id) + API_URL_AFTER)

    if response.status_code == 200:
        json_data = response.json()
        df = pd.DataFrame([json_data])  # Create DataFrame from the relevant data
        dfs.append(df)
    elif response.status_code == 404:
        print(f"Movie with ID {movie_id} not found. Skipping...")
    else:
        print(f"Error in request for movie {movie_id}: {response.status_code}")

# Concatenate all DataFrames into a single one
final_df = pd.concat(dfs, ignore_index=True)

# Now you can work with the final DataFrame
print(final_df.head())
```

```python
def format_string(input_str):
    # Check if '{' is the first character, if not, add it
    if not input_str.startswith('{'):
        input_str = '{' + input_str

    # Check if '}' is the last character, if not, add it
    if not input_str.endswith('}'):
        input_str = input_str + '}'

    return input_str
```

```python
def convert_json_data_to_array(json_data):
    json_string = json.dumps(json_data)
    # Convert JSON to string
    json_string = json_string.replace('[', '')
    json_string = json_string.replace(']', '')
    split_results = json_string.split('}, {')
    result = []
    for split_result in split_results:
        result.append(format_string(split_result))
    return result
```

```python
df_cast = pd.DataFrame(columns=['movie_id', 'cast'])
df_crew = pd.DataFrame(columns=['movie_id', 'crew'])
```

```python
for n in range(len(final_df)):
    movie_id = final_df['id'][n]
    cast_string_array = final_df['cast'][n]
    
    new_cast = {'movie_id': movie_id, 'cast': convert_json_data_to_array(cast_string_array)}
    df_cast = df_cast.append(new_cast, ignore_index=True)
    
    crew_string_array = final_df['crew'][n]
    
    new_crew = {'movie_id': movie_id, 'crew': convert_json_data_to_array(crew_string_array)}
    df_crew = df_crew.append(new_crew, ignore_index=True)
```
